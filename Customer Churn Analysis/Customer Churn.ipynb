```python
df5 = pd.merge(df1,df2,on='CustomerId', how='outer')
```

```python
df6 = pd.merge(df5,df3,on='CustomerId', how='outer')
df6
```

```python
df7 = pd.merge(df6,df4,on='CustomerId', how='outer')
df7
```

```python
df8 = df7.set_index('CustomerId')
df8
```

```python
df8.dropna(how='all',inplace=True)
df8
```

```python
df8.sort_index(inplace=True)
```

```python
df8.reset_index(inplace=True)
df8
```

```python
df8.nunique()
```

We can see that CustomerId is a unique value for each row. The total rows are 10002 and unique CustomerId's are 10000 in number. Hence 2 rows are duplcated. Lets drop duplicate rows

```python
df8.drop_duplicates(inplace=True)
```

```python
df8
```

```python
df8.isnull().sum()
```

```python
null_det = df8.isnull().sum()
sns.barplot(x=null_det.index,y=null_det)
plt.xticks(rotation=45)
```

We can see only 2 columns are have null values. Lets us replace them with median values

```python
df8.shape
```

```python
mcols = ['CreditScore', 'Balance']
for x in mcols:
    df8.loc[df8[x].isnull(),x] = df8[x].median()
    
#df8.fillna(method='ffill',inplace=True)
```
