```python
df5.loc[df5['minutes'].isnull(),'minutes'] = df5[df5['minutes'].isnull()].apply(lambda row:tran3.loc[(row['app_name'],row['Publisher']),'minutes'],axis=1)
```

```python
df5.isnull().sum()
```

We can see that we have replaced the null values in minutes column using the median values we recieved from the intermediate table 'tran3' we recieved after groupby transformation. We can do similar thing for gender_id column also with 1 line of code. We can also observed that the mode value is 2 for all the rows in the intermediate table 'tran2' we recieved after groupby transformation.

```python
df5[df5['gender_id'].isnull()].apply(lambda row:tran2.loc[(row['app_name'],row['Publisher']),'gender_id'],axis=1)
```

```python
df5.loc[df5['gender_id'].isnull(),'gender_id'] = df5[df5['gender_id'].isnull()].apply(lambda row:tran2.loc[(row['app_name'],row['Publisher']),'gender_id'],axis=1)
```

```python
df5.isnull().sum()
```

Now there are no null values in our dataset

Let's perform some filtering based on minutes column if the value is greater than (median + standard deviation) across app_name or minutes value is less than 30 mins. Lets do a groupby across app_name column

```python
tran4 = df5.groupby('app_name').agg({'minutes': ('median', 'std') })
```

```python
tran4
```

```python
tran4 = tran4.droplevel(0,axis=1)
```

```python
tran4['thres'] = tran4['median'] + tran4['std']
```

```python
tran4
```

```python
tran4.loc['Amazon Mobile (Mobile App)','thres']
```

```python
filter1 = df5.apply((lambda row: True if (row['minutes'] <= tran4.loc[row['app_name'],'thres'] and row['minutes'] > 30) else False),axis=1)
```

```python
df6 = df5[filter1]
df6
```

We created a filter by applying lambda function row-wise which will check for the condition using values from the tran4 table. We applied that filter on the df5 dataset to get our output

Lets perform some aggregation on df6 and create a statistical dataset

```python
def Quartile1(x):
    return np.percentile(x,25)
def Quartile3(x):
    return np.percentile(x,75)
```

```python
#df6.groupby('app_name')['minutes'].agg({'count', 'mean', 'min', 'max', 'median', 'std',lambda x: np.percentile(x,25),lambda x: np.percentile(x,75)})
df7 = df6.groupby('app_name')['minutes'].agg({'count', 'mean', 'min', 'max', 'median', 'std',Quartile1, Quartile3})
df7
```

```python
df8 = df6.groupby(['app_name','gender_id'])['minutes'].agg({'count', 'mean', 'min', 'max', 'median', 'std'})
df8
```

Lets create a new dataframe with columns as app_name, “total_minutes”, “total devices”, “Average Minutes per Device Per App”.

```python
df9 = pd.DataFrame()
```

```python
df9['app_name'] = df6.groupby('app_name')['device_id'].count().index
```

```python
df9['total minutes'] = df6.groupby('app_name')['minutes'].sum()
```

```python
s2 = df6.groupby('app_name')['minutes'].sum()
s1= pd.Series(df6.groupby('app_name')['device_id'].count().index)
s3 = df6.groupby('app_name')['device_id'].count()
```

```python
df9 = pd.concat([s2,s3],axis=1)
df9
```

```python
df9.reset_index(inplace=True)
df9
```

```python
df9.columns = ['app_name', 'minutes', 'device_count']
df9.columns
```

```python
df9['Avg_time_spent_per_device'] = df9['minutes']/df9['device_count']
```

```python
df9
```

Let's include a rank column based on minutes, users and publisher

```python
df9['Rank(minutes based)'] = df9['Avg_time_spent_per_device'].rank(ascending=False)
```

```python
df9['Rank(user based)'] = df9['device_count'].rank(ascending=False)
```
